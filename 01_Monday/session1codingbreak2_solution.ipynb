Load all packages you need.

```python
# load libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sweetviz as sv
```

Read the data.

```python
df = pd.read_csv('../data/diabetes.csv')
```

Create an overview of the dataframe (variable names and types; number of observations). What is the target variable? 

```python
df.info()
```

Create a histogram of the variable "BloodPressure".

```python
# Set Seaborn style
sns.set_style("whitegrid")

# Create histogram
plt.figure(figsize=(8, 6))
sns.histplot(df['BloodPressure'], bins=20, kde=True, color='royalblue')
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")
plt.title("Distribution of Blood Pressure")
plt.show()
```

Create a scatterplot of the variables "BloodPressure" and "BMI"

```python
# Create scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['BloodPressure'], y=df['BMI'], color='royalblue', alpha=0.6)
plt.xlabel("Blood Pressure")
plt.ylabel("BMI")
plt.title("Scatterplot of Blood Pressure vs BMI")
plt.show()
```

Run sweetviz on the dataset and inspect it.

```python
report = sv.analyze(df, target_feat="Outcome")
report.show_html("sweetviz_report.html")  # Opens in browser
```
