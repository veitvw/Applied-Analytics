#### Task A
Create a Data Audit for the Pima Indians Diabetes Database. The audit should cover:


    1. Data Overview & Structure:
			- Provide a summary of the dataset.
			- Identify the number of rows and columns.
			- List all variables names and their data types.
			- Classify each variable as categorical or numerical.	

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

from itertools import combinations 
from sklearn.model_selection import train_test_split, KFold
from sklearn.utils import resample

```

```python
df = pd.read_csv('../data/diabetes.csv')
print(df.columns.tolist())
```

```python
# Print the number of rows and columns
print("Number of rows and columns:", df.shape)  # shape returns (rows, columns)

# List all variable names and their data types
print("\nVariable names and their data types:")
print(df.dtypes)  # Displays column names with their data types

# Classify each variable as categorical or numerical
categorical_vars = df.select_dtypes(include=['object']).columns.tolist()  # Identify categorical variables
numerical_vars = df.select_dtypes(include=['number']).columns.tolist()  # Identify numerical variables

print("\nCategorical variables:", categorical_vars)
print("Numerical variables:", numerical_vars)
```

    2. Summary Statistics:
			- Compute key statistics such as mean, median, minimum, maximum, and standard deviation.
			- Create a histogram for each variable.
   			- Identify any unusual values that stand out and possible skew of the data.

```python
# Compute key statistics for numerical variables
print("\nKey statistics for numerical variables:")
stats = df[numerical_vars].describe().T  # Transpose to make it more readable
stats["median"] = df[numerical_vars].median()  # Add median column
print(stats)  # Print statistics including mean, median, min, max, std deviation

# Create histograms for each numerical variable using Seaborn
print("\nGenerating histograms...")
plt.figure(figsize=(12, 8))  # Set figure size
for col in numerical_vars:
    plt.figure(figsize=(6, 4))  # Set individual plot size
    sns.histplot(df[col], kde=True, bins=30)  # Histogram with KDE overlay
    plt.title(f"Distribution of {col}")  # Set title
    plt.xlabel(col)  # Label x-axis
    plt.ylabel("Count")  # Label y-axis
    plt.show()  # Show the plot
```

    3. Outlier:
			- Decide on a formal method to detect univariate outliers.
			- Report the number of outliers per variable.
			- Create scatterplots for each combination of variables.
			- Can you identify possible multivariate outliers that you did not detect as univariate outliers?

```python
# Identify columns where 0 represents missing values
missing_value_columns = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]  # Example columns
# Replace 0 with NaN in specified columns
df[missing_value_columns] = df[missing_value_columns].replace(0, pd.NA)

# Identify columns where 0 represents missing values
missing_value_columns = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]  # Example columns

# Replace 0 with NaN in specified columns
df[missing_value_columns] = df[missing_value_columns].replace(0, pd.NA)

# Identify skewed numerical variables
numerical_vars = df.drop("Outcome", axis=1).select_dtypes(include=['number']).columns.tolist()
skewed_vars = df[numerical_vars].apply(lambda x: x.skew()).sort_values(ascending=False)
skewed_vars = skewed_vars[skewed_vars > 0.5].index.tolist()  # Select highly skewed variables

# Apply log transformation to skewed variables and add as new variables
for var in skewed_vars:
    df[f"log_{var}"] = np.log1p(df[var])  # log1p avoids log(0) issues

# Print transformed variables
print("\nLog transformation applied to skewed variables:", skewed_vars)

# Plot histograms of transformed variables
print("\nGenerating histograms for transformed variables...")
for var in skewed_vars:
    plt.figure(figsize=(6, 4))  # Set individual plot size
    sns.histplot(df[f"log_{var}"], kde=True, bins=30)  # Histogram with KDE overlay
    plt.title(f"Histogram of log-transformed {var}")  # Set title
    plt.xlabel(f"log_{var}")  # Label x-axis
    plt.ylabel("Count")  # Label y-axis
    plt.show()  # Show the plot

print("\nNot great, but maybe good enough for outlier detection...")

# Outlier detection using Tukey's Fence and Z-score methods
outlier_Tukey = []
outlier_z_score = []

for col in numerical_vars:
    if col in skewed_vars:
        col_to_check = f"log_{col}"  # Use log-transformed data for skewed variables
    else:
        col_to_check = col  # Use original data for non-skewed variables
    
    # Tukey's Fence method
    Q1 = df[col_to_check].quantile(0.25)
    Q3 = df[col_to_check].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_Tukey.append((df[col_to_check] < lower_bound) | (df[col_to_check] > upper_bound))
    
    # Z-score method
    mean = df[col_to_check].mean()
    std_dev = df[col_to_check].std()
    outlier_z_score.append(abs(df[col_to_check] - mean) > 3 * std_dev)

# Combine outlier flags into new columns
df["outlier_Tukey"] = np.any(outlier_Tukey, axis=0)
df["outlier_z_score"] = np.any(outlier_z_score, axis=0)

# Define the variables to include in the scatterplot matrix
selected_vars = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Create a copy for visualization
plot_data = df[selected_vars].copy()

# Correct the order of outlier classification to ensure correct labeling
plot_data["Outlier_Label"] = "Normal"
plot_data.loc[(df["outlier_Tukey"] == True) & (df["outlier_z_score"] == True), "Outlier_Label"] = "Both"
plot_data.loc[(df["outlier_Tukey"] == True) & (df["outlier_z_score"] == False), "Outlier_Label"] = "Tukey Only"
plot_data.loc[(df["outlier_Tukey"] == False) & (df["outlier_z_score"] == True), "Outlier_Label"] = "Z-Score Only"

# Define color palette
palette = {"Normal": "black", "Tukey Only": "orange", "Z-Score Only": "pink", "Both": "red"}

# Generate scatterplot matrix using Seaborn with outlier categories as colors
sns.pairplot(plot_data, hue="Outlier_Label", palette=palette, diag_kind="kde", corner=True)
plt.show()
```

    4. Missing Values:
        - Identify columns with missing values.
        - Discuss possible strategies to handle the missing data.

```python
# Hot-deck imputation: Replace NA by randomly sampling from the same column
for column in missing_value_columns:
    df[column] = df[column].apply(lambda x: np.random.choice(df[column].dropna()) if pd.isna(x) else x)
```

    5. Correlations.
			- Create a correlation matrix for the dataset.
			- Visualize the correlation matrix.

```python
# Define variables for correlation analysis
correlation_vars = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Compute the correlation matrix
correlation_matrix = df[correlation_vars].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Selected Variables")
plt.show()
```

Test different methods to sample training and test data from the dataset. Compare the sampled datasets based on the descriptive methods you have learned and evaluate which sampling method worked best.

```python
# Sampling Methods

# 1) Random Sampling
train_random, test_random = train_test_split(df2, test_size=0.25, random_state=42)

# 2) K-Fold Sampling (Creating Fold Indices)
kf = KFold(n_splits=4, shuffle=True, random_state=42)
kf_splits = [(train_idx, test_idx) for train_idx, test_idx in kf.split(df2)]

# 3) Bootstrapping (Sampling with Replacement)
train_bootstrap = resample(df2, replace=True, n_samples=int(0.75 * len(df2)), random_state=42)
test_bootstrap = df2.drop(train_bootstrap.index)

# Print sample sizes
print("Random Sampling: Train size =", len(train_random), ", Test size =", len(test_random))
print("K-Fold Sampling: Number of folds =", len(kf_splits))
print("Bootstrapping: Train size =", len(train_bootstrap), ", Test size =", len(test_bootstrap))
```
