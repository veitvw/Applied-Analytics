We want to perform PCA on these two variables. But before we can do that, we need to normalize the data to avoid weighting the higher "BloodPressure" stronger than "BMI" values. If a variable is normally distributed (bell curve), a good way approach is z-score scaling. <br>
Inspect the histograms of both variables. If they are normally distributed create a dataframe with the normalized "BloodPressure_norm" and "BMI_norm". 

```python
import seaborn as sns
import matplotlib.pyplot as plt

def plot_histogram(df, variable):
    plt.figure(figsize=(8, 6))
    sns.histplot(df[variable], kde=True)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title(f"Histogram of {variable}")
    plt.show()
```

```python
plot_histogram(data, 'BMI')
plot_histogram(data, 'BloodPressure')
```

```python
import pandas as pd
from scipy.stats import zscore

def normalize_variables(df, variables):
    normalized_df = df.copy()
    for var in variables:
        normalized_df[f"{var}_norm"] = zscore(df[var])
    return normalized_df[[f"{var}_norm" for var in variables]]
```

```python
df_norm = normalize_variables(data, ['BMI', 'BloodPressure'])
```

Perform a Principal Component Analyis (PCA) on the two variables and print the eigenvalues, the percentage of the explained variance, and the eigenvectors of the the PCs

```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
def perform_pca(df, variables):
    # Standardizing the data
    standardized_data = (df[variables] - df[variables].mean()) / df[variables].std()

    # Performing PCA
    pca = PCA()
    principal_components = pca.fit_transform(standardized_data)

    # Printing Eigenvalues
    print("Eigenvalues:", pca.explained_variance_)

    # Printing Explained Variance Ratio
    print("Explained Variance Ratio (%):", pca.explained_variance_ratio_ * 100)

    # Printing Eigenvectors
    print("Eigenvectors (Principal Components):")
    print(pca.components_)

    # Creating a DataFrame for Principal Components
    pca_df = pd.DataFrame(principal_components, columns=[f"PC{i+1}" for i in range(len(variables))])

    return pca_df
```

```python
df_pca = perform_pca(df_norm, ['BMI_norm', 'BloodPressure_norm'])
```

Print the eigenvectors on the scatterplot.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def pca_scatterplot_with_eigenvectors(df, variables):
    # Standardizing the data
    standardized_data = (df[variables] - df[variables].mean()) / df[variables].std()

    # Performing PCA
    pca = PCA()
    pca.fit(standardized_data)
    principal_components = pca.transform(standardized_data)

    # Scatter plot of the original data
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=standardized_data[variables[0]], y=standardized_data[variables[1]], alpha=0.6)

    # Plot eigenvectors
    mean_vals = standardized_data.mean().values
    for i in range(len(variables)):
        eigenvector = pca.components_[i]  # Principal component
        eigenvalue = pca.explained_variance_[i]  # Corresponding eigenvalue
        plt.arrow(mean_vals[0], mean_vals[1], 
                  eigenvector[0] * eigenvalue, eigenvector[1] * eigenvalue, 
                  color='red', width=0.02, head_width=0.1, label=f"PC{i+1}")

    # Labels and title
    plt.xlabel(variables[0])
    plt.ylabel(variables[1])
    plt.title("PCA Scatterplot with Eigenvectors")
    plt.legend()
    plt.show()
```

```python
pca_scatterplot_with_eigenvectors(df_norm, ['BMI_norm', 'BloodPressure_norm'])
```
