```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

def normalize_dataframe(df: pd.DataFrame, exclude_columns: list) -> pd.DataFrame:
    scaler = StandardScaler()
    columns_to_normalize = [col for col in df.columns if col not in exclude_columns]
    
    # Normalize the selected columns and add '_norm' to their names
    df_norm = df.copy()
    df_norm[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])
    
    # Rename columns that were normalized
    df_norm = df_norm.rename(columns={col: f"{col}_norm" for col in columns_to_normalize})
    
    return df_norm
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def plot_screeplot(df: pd.DataFrame) -> None:
    # Perform PCA without specifying the number of components to get the explained variance
    pca = PCA()
    pca.fit(df)
    
    # Get the explained variance ratio for each principal component
    explained_variance = pca.explained_variance_ratio_
    
    # Plotting the scree plot
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='--')
    plt.title('Scree Plot')
    plt.xlabel('Principal Components')
    plt.ylabel('Explained Variance Ratio')
    plt.xticks(range(1, len(explained_variance) + 1))
    plt.grid(True)
    plt.show()
```

```python
plot_screeplot(df_norm)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def perform_pca(df_norm: pd.DataFrame, n_components: int = 3) -> pd.DataFrame:
    # Perform PCA on the already normalized data
    pca = PCA(n_components=n_components)
    pca_result = pca.fit_transform(df_norm)
    
    # Create a DataFrame with the PCA results
    pca_df = pd.DataFrame(pca_result, columns=[f"PC{i+1}" for i in range(n_components)])
    
    # Visualize the eigenvalues (components)
    eigenvalues = pca.explained_variance_
    components = pca.components_
    
    # Create a DataFrame for easy visualization of eigenvalues
    eigenvalue_matrix = pd.DataFrame(components.T, columns=[f'PC{i+1}' for i in range(components.shape[0])],
                                     index=df_norm.columns)
    
    # Plot the heatmap with eigenvalues
    plt.figure(figsize=(12, 6))
    sns.heatmap(eigenvalue_matrix, annot=True, cmap="YlGnBu", cbar_kws={'label': 'Eigenvalue'}, fmt='.2f', linewidths=0.5)
    
    plt.title('Eigenvalues of Principal Components across Original Dimensions')
    plt.xlabel('Principal Components')
    plt.ylabel('Original Dimensions (Columns)')
    plt.show()
    
    return pca_df
```

```python
perform_pca(df_norm, 3)
```
